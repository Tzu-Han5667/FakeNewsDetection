## Read data

```python
# China
```

```python
import pandas as pd
def file_reader(path):
    df=pd.DataFrame()
    chunksize = 10 ** 6
    for chunk in pd.read_csv(path, chunksize=chunksize, index_col=[0]):
        df = df.append(chunk)
    return df

country='China'
data=file_reader('/Volumes/BECCA/Project/Data/Input/02_Preprocessing/04_Tweets/%s.csv' %country)
```

```python
data
```

## Hashtags

```python
#preprocessing for hashtag column
import re
for i in range(len(data)):
    data.tweet_text[i]=str(data.tweet_text[i])
# data['tweet_text']=str(data['tweet_text'])
data['hashtag'] = data['tweet_text'].apply(lambda x: re.findall(r"#(\w+)", x))
```

```python
data
```

### Repeated

```python
def unique_list(l):
    l=set(l)
    l=list(l)
    return l


data['repeat_tag']=0
for j in range(len(data)):
    h0=list(data.hashtag[j])
    h1=unique_list(h0)
    if len(h0) != len(h1):
        data['repeat_tag'][j]= True
    elif len(h0) == len(h1):
        data['repeat_tag'][j]= False
```

```python
data
```

### FPM

```python
import pandas as  pd
import numpy as np
import pyfpgrowth
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from collections import Counter

ut=[]
data['unique_tag']=0
for i in range(len(data)):
    l=unique_list(data['hashtag'][i])
    ut.append(l)
data['unique_tag']=ut

tag_list=[]
for i in data['unique_tag']:
    for j in i:
        if j != "":
            tag_list.append(j)

tag_category=unique_list(tag_list)

##count frequency based how many instance tags appear
tag=[]
count=[]
for i in tag_category:
    tag.append(i)
    count.append(tag_list.count(i))
tags_count=np.column_stack((tag, count))

dataset=list(data.unique_tag)

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

col=[]
for i in frequent_itemsets.itemsets:
    for j in i:
        print(j)
        col.append(j)
if len(col)!=0:
    for i in col:
        data[i]=0
        for j in range(len(data)):
            if i in data.hashtag[j]:
                data[i][j]=True
            else:
                data[i][j]=False
```
