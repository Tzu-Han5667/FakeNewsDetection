```python
#China
Country='China'
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('/Volumes/BECCA/Project/Data/Result/%s/By_classifier/Total_Result.csv' %Country, index_col=[0])
```

```python
data
```

# By classifier

## #kNN

```python
clf_name='knn'
result=data.loc[data['Classifier'] == clf_name]
result=result.drop(['Classifier'], axis=1)
```

```python
result
```

```python
plt.figure(figsize=(8,5))

x=result.Combination
for col in result.columns[5:10]:
    y=result[col]
    plt.plot(x,y, label=col)

plt.legend()

plt.savefig('/Volumes/BECCA/Project/Data/Result/%s/By_classifier/%s.png' %(Country,clf_name))
```

```python
opt={'Classifier':clf_name}
for i in list(result.index):
    for j in result.columns:
        if j != 'Combination':
            if result[j].max() == result[j][i]:
                opt[j]=result.Combination[i]
```

```python
Ana_result=pd.DataFrame(opt, index=[0])
Ana_result
```
