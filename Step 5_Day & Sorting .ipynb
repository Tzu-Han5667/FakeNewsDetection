## Read data

```python
# China
```

```python
import pandas as pd
def file_reader(path):
    df=pd.DataFrame()
    chunksize = 10 ** 6
    for chunk in pd.read_csv(path, chunksize=chunksize, index_col=[0]):
        df = df.append(chunk)
    return df

country='China'
data=file_reader('/Volumes/BECCA/Project/Data/Input/02_Preprocessing/04_Tweets/%s.csv' %country)
```

```python
data
```

## Day

```python
from datetime import datetime
hour=[]
minute=[]
day=[]
for i in range(len(data)):
    if len(str(data.tweet_time[i])) < 16:
        data=data.drop([i],axis=0)
    elif len(str(data.tweet_time[i])) == 16:
        j=datetime.strptime(data.tweet_time[i],'%Y-%m-%d %H:%M')
        hour.append(j.hour)
        minute.append(j.minute)
        day.append(j.weekday())
    elif len(str(data.tweet_time[i])) == 19:
        j=datetime.strptime(data.tweet_time[i],'%Y-%m-%d %H:%M:%S')
        hour.append(j.hour)
        minute.append(j.minute)
        day.append(j.weekday())
```

```python
data['day']=day
data['hour']=hour
data['minute']=minute
for i in data.day:
    if i == 0:
        i=7
```

```python
data
```

```python
df=data[['tweet_text','polarity','polarity_number','subjectivity','subjectivity_number','repeat_tag','郭文贵','day','hour','minute','suspend']]
```

```python
df=df.rename(columns={'郭文贵':'freq_tag'})
df
```

```python
df.to_csv('/Volumes/BECCA/Project/Data/Input/02_Preprocessing/05_Final/%s.csv' %country) 
```
