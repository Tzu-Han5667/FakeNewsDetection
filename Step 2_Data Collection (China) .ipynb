## Data collection via Twitter api

```python
keywords=[['milesguo', 'wengui'], ['milesguo', 'china'], ['world', 'epidemic']]

search=[]
for i in keywords:
    i="+".join(i)
    search.append(i)
    
search
```

```python
#Twitter api setting
import os
import tweepy as tw

consumer_key='consumer_key'
consumer_secret='consumer_secret'
access_token='access_token'
access_token_secret='access_token_secret'

auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
#Search Function (first day)
#search word (keywords[i,0:2])
def unique_list(l):
    l=set(l)
    l=list(l)
    return l

import pandas as pd
text=[]
for i in search:
    results=tw.Cursor(api.search, q=i, lang="en").items(2000)
    for tweet in results:
        l=[]
        l.append(tweet.text)
        l.append(tweet.created_at)
        text.append(l)
text=[list(x) for x in set(tuple(x) for x in text)]
```

```python
text=pd.DataFrame(text,columns=['tweet_text', 'tweet_time'])
```

```python
text['suspend']=False
```

## Data combination

```python
import pandas as pd
import os

country='China'

# os.listdir(u"/Users/becca/Desktop/daily")
Folder_Path = u"/Volumes/BECCA/Project/Data/Input/02_Preprocessing/02_Tweet collection"  
SaveFile_Path =u"/Volumes/BECCA/Project/Data/Input/02_Preprocessing/03_Current tweet"
SaveFile_Name = u'%s.csv' %country  

file_list = os.listdir(Folder_Path)

for i in range(len(file_list)):
    if country in file_list[i]:
        df = pd.read_csv(Folder_Path + '/' + file_list[i],usecols=['tweet_text','tweet_time','suspend'])
        df.to_csv(SaveFile_Path + '/' + SaveFile_Name, index=False, header=True, mode='a+')
```

```python
text=pd.read_csv(SaveFile_Path + '/' + SaveFile_Name)
text=text.drop_duplicates(keep='first', inplace=False)
text=text.reset_index()
text=text.drop(['index'],axis=1)
text.to_csv(SaveFile_Path + '/' + SaveFile_Name)
```
