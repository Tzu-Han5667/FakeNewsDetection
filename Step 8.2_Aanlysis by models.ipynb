```python
#China
Country='China'
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('/Volumes/BECCA/Project/Data/Result/%s/By_model/Total_Result.csv' %Country, index_col=[0])
```

```python
data
```

# By feature

## #All

```python
ft_name='k7'
result=data.loc[data['Model'] == ft_name]
result=result.drop(['Model'], axis=1)
```

```python
result
```

```python
plt.figure(figsize=(8,5))

x=result.Classifier
for col in result.columns[5:10]:
    y=result[col]
    plt.plot(x,y, label=col)

plt.legend()
plt.savefig('/Volumes/BECCA/Project/Data/Result/%s/By_model/%s.png' %(Country, ft_name))
```

```python
opt={'Feature Selection':ft_name}
for i in list(result.index):
    for j in result.columns:
        if j != 'Classifier':
            if result[j].max() == result[j][i]:
                opt[j]=result.Classifier[i]
```

```python
Ana_result=pd.DataFrame(opt, index=[0])
Ana_result
```
